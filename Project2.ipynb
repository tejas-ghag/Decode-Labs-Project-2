```python
print("Average Text Length:")
print(df["text_length"].mean())

print("Average Word Count:")
print(df["word_count"].mean())

print("Average Polarity:")
print(df["polarity"].mean())

print("Average Subjectivity:")
print(df["subjectivity"].mean())
```
